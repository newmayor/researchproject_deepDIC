```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '/home/numairahmed/forked_deepDIC/Deep-Dic-deep-learning-based-digital-image-correlation/experiment_numairData/log/log2022_02_05_10:44:50'
log_data = np.loadtxt(path, skiprows=1)
print(log_data)
```

```python
loss_list = []
for ind in range(450):
    index = log_data[:,0]==ind
    loss = log_data[index]
    avg_loss = sum(loss[:,2])/loss[-1,1]
    loss_list.append(avg_loss)

```

```python
plt.figure()
plt.plot(loss_list)
# plt.ylim([0,20])
plt.title('Training loss over epochs')
plt.xlabel('epoch')
plt.ylabel('MSE loss')
plt.show()
```
