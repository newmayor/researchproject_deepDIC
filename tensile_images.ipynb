```python
import numpy as np
import matplotlib.pyplot as plt

from imageio import imread
```

```python
ab = imread('/home/numairahmed/deepDIC/images/image_sample_pairs/experiment_samples/tensile_test_images_2/image_701.png')

import glob

path = '/home/numairahmed/deepDIC/images/image_sample_pairs/experiment_samples/tensile_test_images'
saveto_path = '/home/numairahmed/deepDIC/images/image_sample_pairs/experiment_samples/tensile_test_images/cropped'


w,h = (1,1)
dpi=256


images=glob.glob(path + "/*.bmp")
images = sorted(images)

for ind, image in enumerate(images):
    img = imread(image)
    padd = 300 - img.shape[1]
    img_padded = np.pad(img, [(0,0), (padd//2, padd//2)], mode='edge')
    corner1 = img_padded.shape[0]//2 - 256//2
    corner2 = img_padded.shape[1]//2 - 256//2
    ab_cropped = img_padded[corner1:corner1+256, corner2:corner2+256]

    fig1 = plt.figure(frameon=False)
    fig1.set_size_inches(w,h)
    ax1 = plt.Axes(fig1, [0., 0., 1., 1.])
    ax1.set_axis_off()
    fig1.add_axes(ax1)
    plt.imshow(ab_cropped, cmap='gray', aspect='auto')
    plt.savefig(f"{saveto_path}/train_image_{ind}.png", dpi=dpi)
    plt.close('all')
    # break

    
```
